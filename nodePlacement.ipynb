```python
import matplotlib.pyplot as plt
```

```python

user = (0, 0)  # User at the origin
node1 = (1.7, 0)  # Node 1 is 1.7m to the right of the user
node2 = (0, 1.3)  # Node 2 is 1.3m above the user
plt.figure(figsize=(6, 6))
plt.plot([user[0], node1[0]], [user[1], node1[1]], 'k-', label='User to Node 1 (1.7m)')
plt.plot([user[0], node2[0]], [user[1], node2[1]], 'k-', label='User to Node 2 (1.3m)')
plt.plot([node1[0], node2[0]], [node1[1], node2[1]], 'k--', label='Node 1 to Node 2 (1m)')
plt.scatter(*user, color='purple', label='User')
plt.scatter(*node1, color='yellow', label='Node 1')
plt.scatter(*node2, color='blue', label='Node 2')
plt.text(user[0], user[1] - 0.15, 'User', fontsize=10, ha='center')
plt.text(node1[0] + 0.05, node1[1], 'Node 1', fontsize=10, ha='left')
plt.text(node2[0] - 0.15, node2[1], 'Node 2', fontsize=10, ha='right')
plt.xlabel('X-axis (meters)')
plt.ylabel('Y-axis (meters)')
plt.xlim(-0.5, 2)
plt.ylim(-0.5, 2)
plt.title('Scene Representation')
plt.legend()
plt.grid(True)
# plt.show()
plt.savefig("./processedData/2025-03-30_10-22-15/day3/nodePlacement/2ndHalfscene3.png")
```
